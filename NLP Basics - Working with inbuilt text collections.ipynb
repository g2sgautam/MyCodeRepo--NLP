<h3> Few Concepts </h3> <br/> <br/>

<h3> Token :</h3> <br/> It is the list of words in a text collection 
    
<h3> Word types :</h3> <br/> It is the list of unique words in the collection
    
<h3> Types : </h3> <br/> It is the list of unique elements in the collection

<h3> To explain it more clearly </h3>

Text3 has 44,764 tokens, this book has only 2,789 distinct words, or “word types.” 
A word type is the form or spelling of the word independently of its specific occurrences in a text—that is, the
word considered as a unique item of vocabulary. 
Our count of 2,789 items will include punctuation symbols, so we will generally call these unique items types instead of word
types.

```python
newsent =" Hello Alpha143 you have been credited with 1000 rupees . Enjoy!!"

from nltk.tokenize import word_tokenize

wrds = word_tokenize(newsent)
```

```python
#Alphabetic characters only 

[ele for ele in wrds if ele.isalpha()]
```

```python
#Alphanumeric characters .  no special characters 

[ele for ele in wrds if ele.isalnum()]
```

```python
#Numbers only

[ele for ele in wrds if ele.isdigit()]
```

```python
# All words containing the sequence 'ee'

[ele for ele in wrds if 'ee' in ele]
```

```python
#Arranging bigram by frequency

txt1 ="Aegis is an excellent institute for data science and all aspiring data scientists. Data Science aspirants love working with data"
txt1 += "Data scientists also love working for big companies"

import collections
import operator

lst =nltk.bigrams(nltk.word_tokenize(txt1))

temp_counter =Counter(lst)

sorted(temp_counter,key=operator.itemgetter(1),reverse=True)

```

```python
sum(temp_counter.values())
```

```python
for key,val in temp_counter.items():
    temp_counter[key] = val/sum(temp_counter.values())
print(temp_counter)    
```
